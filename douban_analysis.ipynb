```python
# 分析豆瓣Top250与上映电影年份的关系
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv(r'D:\安装包\data\douban_movies.csv')
data = data[['电影名称','上映日期','豆瓣评分']]
data['上映日期'] = data['上映日期'].str.extract('(\d{4})').astype(int)
# print(data)
data.to_csv(r'D:\安装包\data\clean_douban.csv',index=False)
year_avg = data.groupby('上映日期')['豆瓣评分'].mean().reset_index()
# print(year_avg)

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(12,4))
plt.plot(year_avg['上映日期'],year_avg['豆瓣评分'],marker='o',color='#e74c3c')
plt.title('豆瓣Top250电影：上映年份与平均评分关系（2024）')
plt.xlabel('上映年份')
plt.ylabel('平均评分')
plt.grid(alpha=0.3)
plt.savefig(r'D:\安装包\data\year_rating_trend.png')
```
